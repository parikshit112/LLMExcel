```python
memory = []
prompt = f"""
You are tasked to answer the users questions from the excel sheet given to you.
Do not make facts of your own and answer solely from the excel sheet given to you.

Here is the excel sheet:
{df.to_string()}"""
# question = "Summarize the headers of this table in 200 words"
print("Hi, How can I help you today?")
while(True):
    for i in memory:
        prompt+=i
    question = input("")
    print(question)
    if(question == "exit"):
        break
    question = f"""Question : {question}\nAnswer : """
    response = get_response_together(prompt = prompt + question)
    question += response
    print(response)
    memory.append(question)
```

```python
prompt = f"""
You are tasked to answer the users questions from the excel sheet given to you.
Do not make facts of your own and answer solely from the excel sheet given to you.

Here is the excel sheet:
{df.to_string()}"""

def get_response(question):
    messages = [{
        "role" : "system", "content" : prompt
    }]
    message = {
        "role" : "user", "content" : question
    }
    messages.append(message)
    response = get_response_openai(messages)
    return response
```
